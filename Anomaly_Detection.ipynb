```python
def encode_statement(statement):
  """
  This function preprocesses a statement and generates its sentence embedding.

  Args:
      statement: The text of the statement.

  Returns:
      A sentence embedding vector.
  """
  preprocessed_text = preprocess_text(statement)

  model_name = "all-mpnet-base-v2"  # Replace with your chosen model
  model = SentenceTransformer(model_name)
  statement_vector = model.encode(preprocessed_text)
  return statement_vector
```

```python
def preprocess_text(text):
    text = text.lower()

    text = re.sub(r'http\S+', '', text)

    text = re.sub(r'[^\w\s]', '', text)

    words = nltk.word_tokenize(text)

    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    return words
```

```python
def get_keywords(statement):
    """
    This function extracts keywords from a statement using NLTK, considering a wider range of POS tags
    suitable for legal documents.

    Args:
        statement (str): The text of the legal statement.

    Returns:
        list: A list of extracted keywords (lowercase).
    """

    words = nltk.word_tokenize(statement.lower())

    expected_tags = ["NN", "NNP", "VB", "VBP",  # Basic nouns, verbs
                     "JJ", "JJS", "JJR",  # Adjectives (comparative/superlative) for specific terms
                     "RB", "RBR", "RBS",  # Adverbs (comparative/superlative) for modifiers
                     "NNS",  # Plural nouns for parties or entities
                     "CD", "SYM",  # Cardinal numbers (dates, amounts) and symbols (e.g., §)
                     "$", "%",  # Currency symbols
                     "JJR", "JJS",]  # Focus on comparative/superlative adjectives for obligations]

    keywords = [word for word in words if nltk.pos_tag([word])[0][1] in expected_tags]

    return keywords
```

```python
def determine_num_topics(statements, max_topics=10):
    """
    Determine the optimal number of topics based on coherence score.

    Args:
        statements: A list of statement text strings.
        max_topics: Maximum number of topics to consider.

    Returns:
        Optimal number of topics.
    """
    preprocessed_statements = [preprocess_text(statement) for statement in statements]

    dictionary = Dictionary(preprocessed_statements)

    corpus = [dictionary.doc2bow(statement) for statement in preprocessed_statements]

    best_coherence_score = -1
    best_num_topics = max_topics

    for num_topics in range(1, max_topics + 1):
        lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)

        coherence_model = CoherenceModel(model=lda_model, texts=preprocessed_statements, dictionary=dictionary, coherence='c_v')
        coherence_score = coherence_model.get_coherence()
        if coherence_score > best_coherence_score:
            best_coherence_score = coherence_score
            best_num_topics = num_topics
    print(f"Best Coherence Score : {best_coherence_score}")

    return best_num_topics
```

```python
def detect_topics_keywords_with_auto_num_topics(statements, num_topics):
    """
    This function detects topics and keywords from a corpus of statements using LDA.

    Args:
        statements: A list of statement text strings.
        num_topics: The number of topics to identify.

    Returns:
        A list of topic keywords, where each element is a list of keywords representing a topic.
    """
    preprocessed_statements = [preprocess_text(statement) for statement in statements]

    dictionary = Dictionary(preprocessed_statements)

    corpus = [dictionary.doc2bow(statement) for statement in preprocessed_statements]

    lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics)

    topic_keywords = []
    for topic_id in range(lda_model.num_topics):
        topic_words = [word for word, prob in lda_model.show_topic(topic_id, topn=5)]
        topic_keywords.append(topic_words)

    return topic_keywords
```

```python
def detect_topics_keywords(statements):
    num_topics = determine_num_topics(statements)
    print(f"\nFinding out the topics using the LDA (Latent Dirichlet Allocation) with {num_topics} topics among the provided set of content...")
    return detect_topics_keywords_with_auto_num_topics(statements, num_topics)
```

```python
def detect_anomalies_hybrid(statements):
    print("\nDetecting anomalies in the statements using keywords and cosine similarity...\n")

    print("Determining the number of topics . . .")
    num_topics = determine_num_topics(statements)

    topic_keywords = detect_topics_keywords_with_auto_num_topics(statements, num_topics)

    print("Encoding each statement")
    statement_vectors = [encode_statement(statement) for statement in statements]

    min_keywords = 2

    anomaly_pairs = []

    for i in range(len(statement_vectors)):
        for j in range(i + 1, len(statement_vectors)):
            all_keywords = [keyword for sublist in topic_keywords for keyword in sublist]

            statement_keywords_i = set(get_keywords(statements[i])) & set(all_keywords)
            statement_keywords_j = set(get_keywords(statements[j])) & set(all_keywords)
            shared_keywords = len(set.intersection(statement_keywords_i, statement_keywords_j))

            if shared_keywords >= min_keywords:
                element_wise_product = np.outer(statement_vectors[i], statement_vectors[j])
                norm_product = np.linalg.norm(statement_vectors[i]) * np.linalg.norm(statement_vectors[j])
                similarity = np.sum(element_wise_product) / (norm_product + 1e-8)

                high_threshold = 0.4
                low_threshold = 0.2

                if similarity < high_threshold and similarity > low_threshold:
                    print(f"Potential anomaly detected with similarity {similarity:.2f} between:")
                    print(f"Statement {i+1}: {statements[i]}")
                    print(f"Statement {j+1}: {statements[j]}\n")
                    anomaly_pairs.append((i, j))
    print(f"Number of anomalies : {len(anomaly_pairs)}")
    print("Anomaly detection completed successfully!\n")
    print("************************************************\n")

    return anomaly_pairs
```

```python
import spacy

# Load the spaCy English model for named entity recognition
nlp = spacy.load("en_core_web_sm")

def extract_legal_labels(sentence):
  """
  Extracts people, organizations, locations, dates, monetary values, and legal concepts from a sentence.

  Args:
      sentence (str): The legal sentence to process.

  Returns:
      dict: A dictionary containing extracted labels and their corresponding values.
  """
  labels = {}

  # Process the sentence using spaCy
  doc = nlp(sentence)

  # Extract entities
  for ent in doc.ents:
    if ent.label_ in ("PERSON", "ORG", "GPE", "DATE", "MONEY"):
      labels[ent.label_.lower()] = ent.text  # Lowercase label for consistency

  # Custom logic for legal concepts (replace with your specific rules)
  legal_concepts = ["breach of contract", "force majeure", "intellectual property"]
  for concept in legal_concepts:
    if concept in sentence.lower():  # Lowercase for case-insensitive matching
      labels["legal_concept"] = labels.get("legal_concept", []) + [concept]

  return labels

# Example usage
sentence = "John Smith from Acme Inc. agrees to pay $1 million to Global Corp. by July 1st, 2024, in case of a breach of contract."
labels = extract_legal_labels(sentence)

print(labels)
```
